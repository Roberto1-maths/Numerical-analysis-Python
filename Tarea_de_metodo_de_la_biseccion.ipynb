#  Otro ejemplo del metodo de la bisección

image.png

La función polinomica $f(x)=3x^3-5x^2-4x+4$ tiene en efecto una raiz entre $[0,1]$, cuando $x=\frac{2}{3}$  el polinomio se anula. Para verificar esto podemos usar el programa visto, con la modifcación de cambiar la función y el intervalo.

```python
import numpy as np
import matplotlib.pyplot as plt
f = lambda x: 3*x**3-5*x**2-4*x+4
tol = 1
a, b = 0, 1
x = np.linspace(-2.1, 2.1, 1000)
# graficamos la funcion f
fig, ax = plt.subplots(1, 1, figsize=(12, 4))
ax.plot(x, f(x), lw=1.5)
ax.axhline(0, ls=':', color='k')
ax.set_xticks([-2, -1, 0, 1, 2])
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$f(x)$', fontsize=18)
# Encuentra la Ráiz porel método de bisección y se visualizan
# los pasos del método
fa, fb = f(a), f(b)
ax.plot(a, fa, 'ko')
ax.plot(b, fb, 'ko')
ax.text(a, fa + 0.5, r"$a$", ha='center', fontsize=18)
ax.text(b, fb + 0.5, r"$b$", ha='center', fontsize=18)
n = 1
while b - a > tol:
    m = a + (b - a)/2
    fm = f(m)
    ax.plot(m, fm, 'ko')
    ax.text(m, fm - 0.5, r"$m_%d$" % n, ha='center')
    n += 1
    if np.sign(fa) == np.sign(fm):
            a, fa = m,fm
    else:
            b, fb = m, fm
    ax.plot(m, fm, 'r*', markersize=10)
    ax.annotate("Aproximación de la raiz  %.3f" % m,
    fontsize=14, family="serif",
    xy=(a, fm), xycoords='data',
    xytext=(-150, +50), textcoords='offset points',
    arrowprops=dict(arrowstyle="->",
    connectionstyle="arc3, rad=-.5"))
    ax.set_title("Método de la Bisección")
    print(m)
```

Como se úede ver en el codigo, esta función tiene una raiz en $[0,1]$ cercas de $x=\frac{2}{3}$. Ahora el número de iteraciones nencesarias para conseguir un eros menor a $10^{-6}$ se puede conseguir modificando la tolerancia e el programa y agregando un contado. haciendo la tolerancia igiual a $10^{-6}$

```python
import numpy as np
import matplotlib.pyplot as plt
r=0
f = lambda x: 3*x**3-5*x**2-4*x+4
tol = 10**-6
a, b = 0, 1
x = np.linspace(-2.1, 2.1, 1000)

# Encuentra la Ráiz porel método de bisección y se visualizan
# los pasos del método
fa, fb = f(a), f(b)
ax.plot(a, fa, 'ko')
ax.plot(b, fb, 'ko')
ax.text(a, fa + 0.5, r"$a$", ha='center', fontsize=18)
ax.text(b, fb + 0.5, r"$b$", ha='center', fontsize=18)
n = 1

while b - a > tol:
    m = a + (b - a)/2
    fm = f(m)
    ax.plot(m, fm, 'ko')
    ax.text(m, fm - 0.5, r"$m_%d$" % n, ha='center')
    n += 1
    if np.sign(fa) == np.sign(fm):
            a, fa = m,fm
    else:
            b, fb = m, fm
    r=r+1
s= 2/3 -m
print(r)
print(np.abs(s))
```
